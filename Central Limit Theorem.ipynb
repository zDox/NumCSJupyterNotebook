## Central Limit Theorem
-> basic statement: The central limit theorem states that if you take sufficiently large samples from a population, the samples' means will be normally distributed, even if the population isn't normally distributed
### example

```python
import numpy as np
import matplotlib.pyplot as plt

N= 20
population_happiness_score = np.random.randint(5, size=(N, N))
flat = population_happiness_score.flatten()      
print(flat)
averages = np.zeros(N)
for i in range(0,N):
    
    for j in range(0,N):
        averages[i]+= population_happiness_score[i][j]
        
    averages[i] /= N
plt.hist(flat, bins = 50, density = True)
plt.show()
```

```python
plt.hist(averages, bins=50, density=True)
plt.title("Distribution of sample means")
plt.xlabel("Average happiness score")
plt.ylabel("Density")
plt.show()
```
